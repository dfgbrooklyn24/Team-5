```python
from pathlib import Path


gradRates = pd.read_csv("/content/drive/MyDrive/JPMCHackathonStuff/JPMCSchoolStuff - Copy of SDP_Graduation_Rates_School_S_2023-02-15.csv")
enrollDemo = pd.read_csv("/content/drive/MyDrive/JPMCHackathonStuff/JPMCSchoolStuff - Copy of 2023-2024 Enrollment & Demographics.csv")
collegeMatr = pd.read_csv("/content/drive/MyDrive/JPMCHackathonStuff/JPMCSchoolStuff - Copy of SDP_College_Matriculation_2021-2022.csv")

gradRates = gradRates[(gradRates["cohort"]=="2018-2019")&(gradRates["SubGroup"]=="All Students")]

enrollDemo = enrollDemo[(enrollDemo["GradeLevel"]=="All Grades")]



gradEnroll = gradRates.merge(enrollDemo, on = 'ULCS Code',how ="left")

gradEnroll_CollegeMatr = gradEnroll.merge(collegeMatr, on = 'ULCS Code',how ="left")

gradEnroll_CollegeMatr.sort_values(by = "ULCS Code", ascending = True, inplace = True)

schoolData = schoolIncomes.merge(gradEnroll_CollegeMatr, on = 'ULCS Code',how ="left")
schoolData.sort_values(by = "Income", ascending = True, inplace = True)

schoolData = schoolData[['Zip Code', 'Publication Name', 'ULCS Code', 'Income',
       'Denom', 'Num', 'Graduation Percent', 'StudentEnrollment', 'IEPCount', 'IEPPCT',
       'NotIEPCount', 'NotIEPPCT', 'FemaleCount', 'FemalePCT', 'MaleCount', 'MalePCT', 'BlackAACount',
      'BlackAAPCT', 'WhiteCount', 'WhitePCT', 'EconDisadvRate', 'OtherRaceCount',
      'OtherRacePCT', 'Number of Graduates', 'First-Fall Matriculation Rate']]


#gradRates.keys()

#schoolData.keys()





schoolData
# Zip Code', 'Publication Name', 'ULCS Code', 'Income',
#        'Denom', 'Num', 'Graduation Percent', 'StudentEnrollment', 'IEPCount', 'IEPPCT',
#        'NotIEPCount', 'NotIEPPCT', 'FemaleCount', 'FemalePCT', 'MaleCount', 'MalePCT', 'BlackAACount',
#        'BlackAAPCT', 'WhiteCount', 'WhitePCT', 'EconDisadvRate', 'OtherRaceCount',
#        'OtherRacePCT', 'Number of Graduates', 'First-Fall Matriculation Rate'])

#gradRates.to_csv('/content/drive/MyDrive/JPMCHackathonStuff/gradRatesOut.csv')
schoolData.to_csv('/content/drive/MyDrive/JPMCHackathonStuff/schoolOut.csv')
```
