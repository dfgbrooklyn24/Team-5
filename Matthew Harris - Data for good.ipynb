```python
import pandas as pd 

newphillyd = pd.read_csv('phillydata.csv')
zip_data = pd.read_csv('shared_data_read_only/Geography-Mapping/ZIP_TRACT_122023.csv')
```

```python
newphillyd['GEO_ID_11'] = newphillyd['GEO_ID'].astype(str).str[-11:]
        
# Ensure data types are consistent for merging
newphillyd[ 'GEO_ID_11'] = newphillyd ['GEO_ID_11']. astype(int)
zip_data[ 'TRACT'] = zip_data['TRACT' ].astype(int)
```

```python
merged_data = pd.merge(newphillyd,zip_data[['TRACT','ZIP']],left_on='GEO_ID_11', right_on='TRACT', how='left')

merged_data = merged_data.drop(['GEO_ID_11', 'TRACT'], axis=1)
merged_data

```

```python
merged_data.to_csv('merged_data', index=False)
```

```python
data = pd.read_csv('merged_data.csv')
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame already loaded with necessary data
# Select metrics
metrics = ['S2201_C01_013E', 'S2201_C03_013E', 'S2201_C03_018E', 'S2201_C05_012E', 'S2201_C05_013E', 'S2201_C01_012E']

# Get the top 5 ZIP codes based on the average value of one metric
top_zip_codes = data.groupby('ZIP')['S2201_C01_013E'].mean().nlargest(5).index

# Filter data for these top ZIP codes
filtered_data = data[data['ZIP'].isin(top_zip_codes)]

# Group by ZIP and calculate the mean for all selected metrics
grouped_data = filtered_data.groupby('ZIP')[metrics].mean()

# Rename metrics for better readability
new_metric_names = {
    'S2201_C01_013E': 'Single female house hold',
    'S2201_C03_013E': 'Single female receiving food stamp',
    'S2201_C03_018E': 'Single female not receivng food stamp',
    'S2201_C05_012E': 'Single male household',
    'S2201_C05_013E': 'Single Female not receiving food stamps',
    'S2201_C01_012E': 'Single male not receiving food stamps'
}
grouped_data.rename(columns=new_metric_names, inplace=True)

# Colors
colors = ['orange', 'black', 'green', 'red', 'blue', 'purple']

# Creating the bar chart
ax = grouped_data.plot(kind='bar', figsize=(14, 8), color=colors)
plt.title('Comparison of Metrics Across Top 5 ZIP Codes in Pennsylvania')
plt.xlabel('ZIP Code')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
