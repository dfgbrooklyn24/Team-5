```python
import pandas as pd


snapdata2 = pd.read_csv('singlesnap_data2.csv')
zipdata = pd.read_csv('shared_data_read_only/Geography-Mapping/ZIP_TRACT_122023.csv')
```

```python

snapdata2['GEO_ID_11'] = snapdata2['GEO_ID'].astype(str).str[-11:]

snapdata2['GEO_ID_11'] = snapdata2['GEO_ID_11'].astype(int)
zipdata['TRACT'] = zipdata['TRACT'].astype(int)
```

```python

merged_df = pd.merge(snapdata2, zipdata[['TRACT', 'ZIP']], left_on='GEO_ID_11', right_on='TRACT', how='left')


merged_df = merged_df.drop(['GEO_ID_11', 'TRACT'], axis=1)
merged_df
```

```python
merged_df.to_csv('philadelphiasnap_data2.0.csv', index=False)
```
